This is a simple Python project built in **Jupyter Notebook** to track daily expenses, 
store them in a CSV file, and visualize spending with charts.

---


## Step 1: Import Libraries

```python

import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

## Step 2: Add Expense Entry

```python

date = datetime.date.today()
category = input("Enter category (Food/Travel/Shopping/Other): ")
amount = float(input("Enter amount: "))
data = pd.DataFrame([[date, category, amount]], columns=["Date", "Category", "Amount"])
data.to_csv("expenses.csv", mode='a', header=False, index=False)

print("âœ… Expense added successfully!")
```

## Step 3: View All Expenses

```python

# Read all saved expenses
df = pd.read_csv("expenses.csv", names=["Date", "Category", "Amount"])
df.head()
```

## Step 4: Summary Report

```python

summary = df.groupby("Category")["Amount"].sum()
summary
```

## Step 5: Visualize Expenses

```python
plt.figure(figsize=(7,5))
summary.plot(kind='bar', color=['#4A90E2','#50E3C2','#F5A623','#D0021B'])
plt.title("Expenses by Category", fontsize=14, fontweight="bold")
plt.xlabel("Category")
plt.ylabel("Total Amount (â‚¹)")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
## Step 6: Visualize Expenses (Pie Chart)
# Pie chart of expenses by category
plt.figure(figsize=(6,6))
plt.pie(summary, labels=summary.index, autopct='%1.1f%%', startangle=140,
        colors=['#4A90E2','#50E3C2','#F5A623','#D0021B'])
plt.title("Expenses Distribution (Pie Chart)")
plt.show()
```
